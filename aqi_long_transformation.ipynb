```python
import pandas as pd
```

```python
df = pd.read_csv('df_fill_cleaned.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
polutan_cols = ['pm10', 'pm25', 'so2', 'co', 'o3', 'no2']
df['max_value'] = df[polutan_cols].max(axis=1)
df['max_polutan'] = df[polutan_cols].idxmax(axis=1)
def kategori_ispu_indonesia(val):
    if val <= 50:
        return "Baik"
    elif val <= 100:
        return "Sedang"
    elif val <= 199:
        return "Tidak Sehat"
    elif val <= 299:
        return "Sangat Tidak Sehat"
    else:
        return "Berbahaya"

df['kategori_ispu'] = df['max_value'].apply(kategori_ispu_indonesia)
df_long = df.melt(
    id_vars=['tanggal', 'year', 'max_value', 'max_polutan', 'kategori_ispu'],
    value_vars=polutan_cols,
    var_name='polutan',
    value_name='nilai'
)
```

```python
df_long.head()
```

```python
df_long.to_csv('df_long.csv', index=False)
```
